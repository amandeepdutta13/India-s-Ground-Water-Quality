```python
#what are the top 10 states with high Mean Temperature
top_states=df.groupby("STATE")['Temperature_Mean'].mean().sort_values(ascending=False).to_frame().reset_index()
top10_states=top_states.head(10)
```

```python
sb.catplot(data=top10_states, x='Temperature_Mean', y='STATE', kind='point',linestyles=["--"])
plt. title( 'Top 10 States with high Mean Temperature')
plt.ylabel( 'States')
plt.xlabel('Temperature (°C)')
```

```python
#Seggregrating the Nature of Water

def condition(row):
    if row['pH_Mean']>7.0:
        return 'Basic'
    elif row['pH_Mean']<7.0:
        return 'Acidic'
    else:
        return 'Neutral'
```

```python
df['Nature_of_Water']=df.apply(condition,axis=1)
df.head(10)
```

```python
acidic_states=df[df['Nature_of_Water'] == 'Acidic']
basic_states=df[df['Nature_of_Water'] == 'Basic']
```

```python
acidic_stations=acidic_states.groupby(['STATE','Year'])['Nature_of_Water'].count().sort_values(ascending=False).to_frame().reset_index()
basic_stations=basic_states.groupby(['STATE','Year'])['Nature_of_Water'].count().sort_values(ascending=False).to_frame().reset_index()
```

```python
g = sb.FacetGrid(data = acidic_stations, col ='STATE', col_wrap=5)
g.map(plt.bar,'Year','Nature_of_Water', color='red', alpha = 0.5, linewidth=.8, edgecolor="black")
g.set_titles("{col_name}")
g.set_xlabels('Year')
g.set_ylabels('Number of Acidic stations')
plt.subplots_adjust(top=0.92)
g.fig.suptitle('Total Number of Stations with Acidic nature across the States throughout the years');
```

```python
g = sb.FacetGrid(data = basic_stations, col ='STATE', col_wrap=5)
g.map(plt.bar,'Year','Nature_of_Water', color='blue', alpha = 0.5, linewidth=.8, edgecolor="black")
g.set_titles("{col_name}")
g.set_xlabels('Year')
g.set_ylabels('Number of Acidic stations')
plt.subplots_adjust(top=0.92)
g.fig.suptitle('Total Number of Stations with Basic nature across the States throughout the years');
```
